```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
print(ratings_with_name['Book-Rating'].dtype)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] > 250].sort_values('avg_ratings',ascending=False)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_ratings','avg_ratings','Image-URL-L','Year-Of-Publication','Publisher']]
popular_df
```

COLLABORATIVE

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
voting_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(voting_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
print(famous_books[:5])
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
sorted_scores = sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)[1:11]
```

```python
df = pd.DataFrame(pt.index.tolist(),columns=['Book_Name'])
df['sorted_scores'] = [sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11] for index in range(pt.shape[0])]
```

```python
def recommend(book_name):
    similar_items = df[df['Book_Name'] == book_name]['sorted_scores'].values[0]
    for i in similar_items:
        book = df.iloc[i[0]]
        print(book)

```
